## <span style="color:#00CED1"><strong>Data Cleaning</strong></span>

### **Finding missing values**

```python
missing_values = df.isnull().sum()

print("Missing values in each column:")
print(missing_values)
```

#### There are **263** missing values in **Age** column, **1014** missing values in **Cabin** column, **1** missing value in **Fare** column and **2** missing values in **Embarked** column.

### **Handling Missing Values**

#### We can replace missing values in **Age** column by **median**. Most of the values in **Cabin** column are NaN values. So, this column can be removed. Rest of the missing values can be dropped.

```python
# Replace missing values in the 'Age' column with the median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Delete the 'Cabin' column
df.drop(columns='Cabin', inplace=True)

# Drop rows with missing values in any other columns
df.dropna(inplace=True)

# Verify no missing values remain
print("Missing values after cleaning:")
print(df.isnull().sum())
```

#### All the missing values are removed.

### **Dropping Irrelevant columns**

#### Passenger_Id is an irrelevant column. It can be removed.

```python
df.drop(columns='Passenger_Id', inplace=True)
```

```python
df.shape
```

#### Now we have 1306 rows and 10 columns.

#### Replace 0 with 'No' and 1 with 'Yes' in the 'Survived' column

```python
df['Survived'] = df['Survived'].replace({0: 'No', 1: 'Yes'})
df.head()
```

### **Handling Outliers**

#### **Box plot for numerical columns**

```python
sns.set(style="whitegrid")

# Create boxplots for each numerical column
plt.figure(figsize=(10, 7))  
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 2, i)  
    sns.boxplot(x=df[column], color='orange')
    plt.title(f'Boxplot of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)

plt.tight_layout()
plt.show()
```

#### **Detecting Outliers**

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]
```

```python
numerical_columns
```

```python
outliers_dict = {}

for column in numerical_columns:
    outliers = detect_outliers_iqr(df, column)
    outliers_dict[column] = outliers
    print(f"Outliers in {column}:")
    print(outliers)
    print(f"Number of outliers in {column}: {len(outliers)}\n")
```

#### **Outlier removal using IQR method**

```python
# Function to remove outliers using IQR
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Remove outliers for multiple columns
for column in numerical_columns:
    df1 = remove_outliers_iqr(df, column)

# Check the shape of the dataset after outlier removal
print(f"Dataset shape after outlier removal: {df1.shape}")
```

#### **Box plot of  umerical columns after outlier removal**

```python
sns.set(style="whitegrid")

# Create boxplots for each numerical column
plt.figure(figsize=(10, 7))  
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 2, i)  
    sns.boxplot(x=df1[column], color='violet')
    plt.title(f'Boxplot of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)

plt.tight_layout()
plt.show()
```
